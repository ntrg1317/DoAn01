```python
import pandas as pd
import numpy as np
```

```python
subject = pd.read_csv("D:/trang/doan/DoAn1/mdpi_crawler/data/staging1/subject.csv")
journal = pd.read_csv("D:/trang/doan/DoAn1/mdpi_crawler/data/staging1/journal.csv")
paper = pd.read_csv("D:/trang/doan/DoAn1/mdpi_crawler/data/staging1/paper.csv")
```

```python
subject['id'] = range(1, len(subject) + 1)

# Move 'Index' column to the first position
subject = subject[['id'] + [col for col in subject.columns if col != 'id']]
subject
```

<h2>Xử lý bảng journal</h2>

```python
journal.replace('\n', '', regex=True, inplace=True)
journal["IF"].replace('-', 0, regex=True, inplace=True)
journal["CiteScore"].replace('-', 0, regex=True, inplace=True)
journal['Upcoming Articles'].replace('N/A', 0, inplace=True)
journal
```

```python
journal.drop(columns=['RSS', 'Current Issue', 'ISSN'], inplace=True)
journal.rename(columns={'#' : 'id'}, inplace=True)
journal['id'] = journal['id'].astype('int64')
journal.columns = [col.lower().replace(' ', '_') for col in journal.columns]
journal.loc[journal['journal_id'] == 'psych', 'journal_id'] = 'ejihpe'

journal
```

```python
journal['upcoming_articles'] = journal['upcoming_articles'].fillna(0)
journal
```

```python
journal.info()
```

```python
journal['upcoming_articles'] = pd.to_numeric(journal['upcoming_articles'], errors='coerce').fillna(0).astype('int64')
journal['total_articles'] = journal['total_articles'].str.replace(',', '').astype('int64')
journal['if'] = pd.to_numeric(journal['if'], errors='coerce').fillna(0).astype(float)
journal['citescore'] = journal['citescore'].astype(float)
```

```python
journal
```

```python
journal.info()
```

<h3>Xử lý bảng paper

```python
paper
```

```python
paper['pubdate'] = paper['pubdate'].str.extract(r'(\d{2} \w+ \d{4})')
paper.loc[paper['journal'] == 'psych', 'journal'] = 'ejihpe'
paper['pubdate'] = pd.to_datetime(paper['pubdate'], format='%d %b %Y')
```

```python
paper
```

```python
paper.info()
```

```python
paper_1 = paper.merge(subject, on='subject', how='left')
paper_1.drop(columns='subject', inplace=True)
paper_1
```

```python
paper = paper_1.groupby(['title', 'author'], as_index=False).agg({
    'id': lambda x: ', '.join(map(str, np.unique(x))),
    'abstract': 'first',
    'journal': 'first',
    'pubdate': 'first'
})

paper.rename(columns={'id': 'subject_id'}, inplace=True)

paper['id'] = paper.index + 1

cols = ['id'] + [col for col in paper.columns if col != 'id']
paper = paper[cols]

paper = paper.drop_duplicates(subset=["title", "author"])

paper.reset_index(drop=True, inplace=True)

paper
```

<h3>Tạp bảng paper_subject</h3>

```python
paper['subject_id'] = paper['subject_id'].apply(lambda x: [a.strip() for a in x.split(',')])
paper_subject = paper.explode('subject_id')[['id', 'subject_id']]

paper_subject.columns = ['id', 'subject_id']
paper_subject['subject_id']= paper_subject['subject_id'].astype('int64')
paper_subject = paper_subject.rename(columns={'id':'paper_id'})
paper_subject
```

```python
paper_subject.info()
```

```python
paper.drop(columns=['subject_id'], inplace=True)
paper
```

<h3>Tạo bảng paper journal</h3>

```python
paper_journal = paper.merge(journal, how='left', left_on='journal', right_on='journal_id')

paper_journal = paper_journal[['id_x', 'id_y']]
paper_journal = paper_journal.rename(columns={'id_x': 'paper_id', 'id_y': 'journal_id'})
paper_journal['journal_id'] = paper_journal['journal_id'].astype('int64')
paper_journal
```

```python
paper.drop(columns="journal", inplace=True)
paper
```

<h3>Tạo bảng author</h3>

```python
paper['author'] = paper['author'].str.replace('by\n', '')

paper['author'] = paper['author'].apply(lambda x: [a.strip() for a in x.replace(' and ', ',').split(',')])

paper
```

```python
# Explode the authors into separate rows
exploded_authors = paper.explode('author')

# Get unique authors
author = exploded_authors['author'].unique()

# Create the author DataFrame with IDs
author = pd.DataFrame({'id': range(1, len(author) + 1), 'author': author})
author
```

```python
merged = exploded_authors.merge(author, on='author')

# Create the paper_author DataFrame
paper_author = merged[['id_x', 'id_y']]
paper_author = paper_author.rename(columns={'id_x': 'paper_id', 'id_y': 'author_id'}).drop_duplicates()

paper_author
```

```python
paper.drop(columns="author", inplace=True)
paper
```
